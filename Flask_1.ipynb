Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans1. Flask is a micro web framework written in Python. It is designed to be lightweight and modular, providing developers with the flexibility to create web applications and APIs quickly and easily. Flask does not impose any particular project structure or dependencies, allowing developers to choose the components they need and build applications according to their specific requirements.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

Ans3.In Flask, app routing refers to the process of mapping URLs to specific functions or view handlers within a Flask application. It allows you to define different routes or URLs that users can access and associate them with the appropriate code that should be executed when those URLs are requested.

Routes in Flask are defined using the @app.route() decorator, which is a way to associate a URL pattern with a Python function. The @app.route() decorator is used to mark a function as a handler for a specific URL route. The decorator takes the URL pattern as its argument and binds it to the function below it.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans5.
In Flask, the url_for() function is used for URL building. It generates a URL for the specified endpoint, taking into account the routing rules defined in the application.

The url_for() function takes the endpoint name as its first argument and optional keyword arguments for any dynamic parts of the URL route. It returns the URL as a string.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def user_profile(username):
    return f"User profile: {username}"

@app.route('/')
def index():
    # Generate the URL for the 'user_profile' endpoint and pass the 'username' argument
    user_url = url_for('user_profile', username='JohnDoe')
    return f"User URL: {user_url}"

if __name__ == '__main__':
    app.run()
```
